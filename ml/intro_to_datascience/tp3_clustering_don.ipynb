```python
import pandas as pd
don_csv = pd.read_csv("./donclassif.csv", sep=';')
don_df = pd.DataFrame(don_csv)
don_df.head()
```

```python
don_df.info()
```

```python
import seaborn as sns
sns.scatterplot(x="V1", y="V2", data=don_df, alpha=0.4)
```

```python
g = sns.PairGrid(don_df)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=10);
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage  
from matplotlib import pyplot as plt

linked = linkage(don_df, 'single')

plt.figure(figsize=(7, 4))  
dendrogram(linked,  orientation='top',distance_sort='descending',show_leaf_counts=True)
plt.show()  
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import davies_bouldin_score

n_clusters = [i for i in range(2,10)]
models_clustering = []
db_score = []
for num_clusters in n_clusters:
    cluster = AgglomerativeClustering(n_clusters=num_clusters, affinity='euclidean', linkage='ward')
    cluster.fit_predict(don_df)
    models_clustering.append(cluster)
    
    db_score.append(davies_bouldin_score(don_df, cluster.labels_))
#     print(num_clusters)
print( db_score)
```

```python
from matplotlib import pyplot as plt

i = 0
plt.figure(figsize=(15, 25))  
for num_clusters in n_clusters:
    plt.subplot(len(n_clusters), 2, i+1)
    plt.scatter(don_df["V1"], don_df["V2"], c=models_clustering[i].labels_, cmap='tab10', marker='.')
    i +=1
plt.show()
```

```python
from sklearn.cluster import KMeans

n_clusters = [i for i in range(2,10)]
kmeans_clustering = []
db_kmeans_score = []
for num_clusters in n_clusters:
    cluster = KMeans(n_clusters=num_clusters, random_state=42)
    cluster.fit(don_df)
    kmeans_clustering.append(cluster)
    
    db_kmeans_score.append(davies_bouldin_score(don_df, cluster.labels_))
#     print(num_clusters)
print( db_kmeans_score)
```

```python
from matplotlib import pyplot as plt

i = 0
plt.figure(figsize=(15, 25))  
for num_clusters in n_clusters:
    plt.subplot(len(n_clusters), 2, i+1)
    plt.scatter(don_df["V1"], don_df["V2"], c=kmeans_clustering[i].labels_, cmap='tab10', marker='.')
    i +=1
plt.show()
```

```python
for i in range(len(db_score)):
    print(db_score[i], db_kmeans_score[i])
```
