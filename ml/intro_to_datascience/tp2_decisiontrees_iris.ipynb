```python
import pandas as pd
import seaborn as sns
import graphviz
```

```python
from sklearn.model_selection import train_test_split
iris_csv = pd.read_csv("./Iris.csv").drop("Id", axis=1)

# iris_x_train, iris_x_test, iris_y_train, iris_y_test = train_test_split(iris_csv.drop("Species", axis=1), iris_csv["Species"], test_size=0.33, random_state=42)
iris_train, iris_test = train_test_split(iris_csv, test_size=0.25, random_state=42)
print(len(iris_train), "train +", len(iris_test), "test")
iris_data_set = pd.DataFrame(iris_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

flower_classifier = DecisionTreeClassifier(max_depth=2, min_samples_split=0.1)
cross_val = cross_val_score(flower_classifier, iris_data_set.drop("Species",axis=1), iris_data_set["Species"] , cv=10)
print(cross_val)
```

```python
flower_classifier.fit(iris_data_set.drop("Species",axis=1).drop("SepalWidthCm",axis=1), iris_data_set["Species"])
flower_classifier.score(iris_test.drop("Species",axis=1).drop("SepalWidthCm",axis=1), iris_test["Species"])
```

```python
prediction = flower_classifier.predict(iris_data_set.drop("Species",axis=1).drop("SepalWidthCm",axis=1))
```

```python
from sklearn import tree
tr = tree.export_graphviz(flower_classifier)
graph = graphviz.Source(tr)
graph
```

```python
from dtreeviz.trees import *
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(["Iris-setosa", "Iris-versicolor", "Iris-virginica"])
target = le.transform(np.array(iris_data_set["Species"]).reshape(-1, 1))
iris_no_sepal = iris_data_set.drop("SepalWidthCm",axis=1).drop("Species",axis=1)

viz = dtreeviz(flower_classifier, 
               iris_no_sepal, 
               target,
               target_name='Species',
               feature_names=list(iris_no_sepal.columns.values), 
               class_names=["Iris-setosa", "Iris-versicolor", "Iris-virginica"]  # need class_names for classifier
              )  
              
viz
```
